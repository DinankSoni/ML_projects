now fill missing values.

```python
# Fill null values with mean for each column
df_filled = df.fillna(df.mean())

# Display the DataFrame with null values filled by mean
df_filled
```

```python
# Find columns with missing values (True indicates missing values)
columns_with_missing_values = df.isnull().sum() > 0

columns_with_missing_values[columns_with_missing_values].index.tolist()
```

```python
# List of specific columns for which you want to find unique values
specific_columns = ['LotFrontage',
 'MasVnrType',
 'MasVnrArea',
 'BsmtQual',
 'BsmtCond',
 'BsmtExposure',
 'BsmtFinType1',
 'BsmtFinType2',
 'Electrical',
 'GarageType',
 'GarageYrBlt',
 'GarageFinish',
 'GarageQual',
 'GarageCond']  

# Initialize a dictionary to store unique values for each specific column
unique_values_dict = {}

# Iterate through the specific columns and find unique values
for column in specific_columns:
    unique_values = df[column].unique()
    mode_value = df[column].mode().iloc[0]  # Get the most common value (mode)
    unique_values_dict[column] = {
        'Unique Values': unique_values,
        'Most Common Value (Mode)': mode_value
    }

# Print unique values for each specific column
for column, values_info in unique_values_dict.items():
    print(f"Unique values for {column}:")
    print(f"Unique Values: {values_info['Unique Values']}")
    print(f"Most Common Value (Mode): {values_info['Most Common Value (Mode)']}\n")
```

```python
# List of columns and their default values for filling
columns_to_fill = {
    'LotFrontage': 60.00,
    'MasVnrType': 'None',
    'MasVnrArea': 400.0,  # we see there are common power is 2 and looks like 4 is the average.
    'BsmtQual': 'TA',
    'BsmtCond': 'TA',
    'BsmtExposure': 'No',
    'BsmtFinType1': 'Unf',
    'BsmtFinType2': 'Unf',
    'Electrical': 'SBrkr',
    'GarageType': 'Attchd',
    'GarageYrBlt': 2005.0,
    'GarageFinish': 'Unf',
    'GarageQual': 'TA',
    'GarageCond': 'TA'
}

# Loop through the specified columns and fill null values with default values
for column, default_value in columns_to_fill.items():
    df[column].fillna(default_value, inplace=True)
```

```python
# Find columns with missing values (True indicates missing values)
columns_with_missing_values = df.isnull().sum() > 0

columns_with_missing_values[columns_with_missing_values].index.tolist()
```

```python
# we need labal encoder
from sklearn.preprocessing import LabelEncoder

# Get a list of column names with non-numeric (categorical) data
categorical_columns = df.select_dtypes(include=['object']).columns

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
y = df.iloc[:,75:]
x = df.drop(columns=['SalePrice'], axis=1)
y,x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
```

```python
x_train.describe()
```

###  here is big differnce in columns values we need StandardScaler

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the entire DataFrame
x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
x_test = pd.DataFrame(scaler.fit_transform(x_test), columns=x_test.columns)
```

```python
x_train.describe()
```

```python
x_test.describe()
```

### reduce column for x

```python
from sklearn.decomposition import PCA

n_components = 5  # Set the number of components you want to keep
pca = PCA(n_components=n_components)

x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
# Create DataFrames for the principal components
principal_train_df = pd.DataFrame(data=x_train_pca, columns=[f'PC{i}' for i in range(1, n_components + 1)])
principal_test_df = pd.DataFrame(data=x_test_pca, columns=[f'PC{i}' for i in range(1, n_components + 1)])
```

```python
principal_train_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and fit the linear regression model
regression_model = LinearRegression()
regression_model.fit(principal_train_df, y_train)

# Predict on the test set
y_pred = regression_model.predict(principal_test_df)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Linear Regression Metrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.linear_model import Ridge

# Initialize and fit the Ridge regression model
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(principal_train_df, y_train)

# Predict on the test set
y_pred = ridge_model.predict(principal_test_df)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Ridge Regression Metrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.linear_model import Lasso

# Initialize and fit the Lasso regression model
lasso_model = Lasso(alpha=1.0)
lasso_model.fit(principal_train_df, y_train)

# Predict on the test set
y_pred = lasso_model.predict(principal_test_df)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Lasso Regression Metrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and fit the Random Forest regression model
rf_model = RandomForestRegressor(n_estimators=200, random_state=42)
rf_model.fit(principal_train_df, y_train)

# Predict on the test set
y_pred = rf_model.predict(principal_test_df)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Random Forest Regression Metrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```
