```python
# List of specific columns for which you want to find unique values
specific_columns = ['MasVnrType', 'BsmtQual', 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical','GarageType','GarageFinish','GarageQual','GarageCond']  

# Initialize a dictionary to store unique values for each specific column
unique_values_dict = {}

# Iterate through the specific columns and find unique values
for column in specific_columns:
    unique_values = df[column].unique()
    unique_values_dict[column] = unique_values

# Print unique values for each specific column
for column, unique_values in unique_values_dict.items():
    print(f"Unique values for {column}:")
    print(unique_values)
```

```python
# List of columns and their default values for filling
columns_to_fill = {
    'MasVnrType': 'BrkFace',
    'BsmtQual': 'TA',
    'BsmtCond': 'TA',
    'BsmtExposure': 'Av',
    'BsmtFinType1': 'Unf',
    'BsmtFinType2': 'Unf',
    'Electrical': 'SBrkr',
    'GarageType': 'Attchd',
    'GarageFinish': 'Unf',
    'GarageQual': 'TA',
    'GarageCond': 'TA'
}

# Loop through the specified columns and fill null values based on unique values
for column, default_value in columns_to_fill.items():
    unique_values = df[column].unique()
    
    # Check if 'None' or 'No' is in unique values, and fill accordingly
    if 'None' in unique_values or 'No' in unique_values:
        df[column] = df[column].fillna(default_value).apply(lambda x: default_value if x in ['None', 'No'] else x)
    else:
        df[column].fillna(default_value, inplace=True)
```
